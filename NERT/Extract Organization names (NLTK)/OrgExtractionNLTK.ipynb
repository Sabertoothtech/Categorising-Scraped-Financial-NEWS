```python
#NER USING NLTK

import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk import ne_chunk, pos_tag
import csv
def entities(text):
    return ne_chunk(
    pos_tag(
    word_tokenize(text)))

file = open(r'moneycontrol.csv','r',encoding='ISO-8859-1')
new = open(r'OrgNames.csv','a',newline='')
read = csv.reader(file)
writ=csv.writer(new)
writ.writerow(['Content','Organization Names'])
next(read, None)
count=0
for reader in read:
    try:
        string=""
        tree = entities(reader[8])
        org = []
        for subtree in tree.subtrees(filter=lambda t: t.label() == 'ORGANIZATION'):
            for leave in subtree.leaves():
                org.append(leave)
        for z in org:
            #print(z[0])
            string = string+z[0]+", "
        print(string[:-2])
        writ.writerow([reader[8],string[:-2]])
        count+=1
        print(count)
    except:
        pass
```
