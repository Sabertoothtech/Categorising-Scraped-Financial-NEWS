# Extract organization names from a text using Stanford Named Entity Recognizer
Download the Stanford Named Entity Recognizer from https://nlp.stanford.edu/software/CRF-NER.html

This is configured to run for Java.

__st = StanfordNERTagger(r'stanford-ner-2018-02-27\classifiers\english.all.3class.distsim.crf.ser.gz',r'stanford-ner-2018-02-27\stanford-ner.jar')__ will configure it to use in Python.

The news file containing the title is given as input to the Entity Recognizer and the Title (on which the NER acts) along with the organization named extracted are given as output.

```python
#NER using STANFORD NER
from nltk.tag.stanford import StanfordNERTagger
from nltk import word_tokenize
import csv
def cleanString(incomingString):
    newstring = incomingString
    newstring = newstring.replace("!","")
    newstring = newstring.replace("@","")
    newstring = newstring.replace("#","")
    newstring = newstring.replace("$","")
    newstring = newstring.replace("%","")
    newstring = newstring.replace("^","")
    newstring = newstring.replace("&","and")
    newstring = newstring.replace("*","")
    newstring = newstring.replace("(","")
    newstring = newstring.replace(")","")
    newstring = newstring.replace("+","")
    newstring = newstring.replace("=","")
    newstring = newstring.replace("?","")
    newstring = newstring.replace("\'","")
    newstring = newstring.replace("\"","")
    newstring = newstring.replace("{","")
    newstring = newstring.replace("}","")
    newstring = newstring.replace("[","")
    newstring = newstring.replace("]","")
    newstring = newstring.replace("<","")
    newstring = newstring.replace(">","")
    newstring = newstring.replace("~","")
    newstring = newstring.replace("`","")
    newstring = newstring.replace(":","")
    newstring = newstring.replace(";","")
    newstring = newstring.replace("|","")
    newstring = newstring.replace("\\","")
    newstring = newstring.replace("/","")
    newstring = newstring.replace(",","")
    newstring = newstring.replace(".","")
    #newstring = newstring.lower()
    return newstring
st = StanfordNERTagger(r'C:\Users\T\Downloads\stanford-ner-2018-02-27\classifiers\english.all.3class.distsim.crf.ser.gz',r'C:\Users\T\Downloads\stanford-ner-2018-02-27\stanford-ner.jar')
file = open(r'C:\Users\T\Desktop\Saber\Filtering the News\NERT\news3.csv','r',encoding='ISO-8859-1')
new = open(r'C:\Users\T\Desktop\Saber\Filtering the News\NERT\OrgNames.csv','a',newline='')
read = csv.reader(file)
writ=csv.writer(new)
writ.writerow(['Title','Organization Names'])
next(read, None)
count=0
for reader in read:
        string=""
        temp = cleanString(str(reader[1]))
        #print(temp)
        tree=st.tag(word_tokenize(temp))
        #print(tree)
        for z in tree:
            if(z[1]=='ORGANIZATION'):
                #print(z[0])
                string = string+z[0]+" "
        print(string)
        writ.writerow([temp,string])
        count+=1
        print(count)
    
new.close()
```
